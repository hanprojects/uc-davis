## Fundamental Terms

An **experiment** is some action with distinct **outcomes**. Each of these outcomes occurs with some (loosely) chance (formally) **probability**. 

**Sample Space**: A sample space denoted by $S$ is the set of all possible outcomes of an experiment.  

Note 1: The term *experiment* has a very general interpretation. It could be  tossing a coin or measuring the protien level in a blood sample to determine malignancy. 

**Non-determinism**: While we know the sample space, i.e., all the outcomes that comprises the sample space, for a specific experiment we do not know what will be the outcome. When we measure the protien level in the blood, we know the range within which the measured value could lie  but we do not know what will be specific value for a given blood sample.

**Event**: An event is a subset of the sample space, and as such, it is a set of outcomes. All the typical properties of set theory still apply here.
Note that an event could be one outcome or a set of outcomes. Don't confuse the idea of "event" with experiment: one experiment could have multiple events apply to its outcome. For example, when drawing a single card, the card could be a Red Card and a Number Card. Both of these could be represented by different events resulting from the same physical experiment. 

Figure \@ref(fig:Sample-Space) shows a sample space $S$. The sample space consists of outcomes represented by the dots. It shows 2 events, Event $A$ and Event $B$. Note that an event can be a single outcome as it is the case for Event $B$. 

![Sample-Space](./Figures/Sample-space-and-events.jpg)

The sample space can be finite, countably infinite, or uncountably infinite. Countably infinite corresponds to the set of integers and uncountably infinite corresponds to the real numbers. We consider examples of each type.

### Finite Sample Space

***Example 1: Tossing a Coin*** What is the sample space for the experiment of tossing a fair coin?  
**Solution:** The sample space $S = \{H, T\}$ where H denotes a Head and T denotes a Tail.

***Example 2: Rolling Dice***  What is the samples space for rolling 2 dice of different colors?  


![Finite-Sample-Space](./Figures/Finite-state-space.jpg)

**Solution:** The sample space is 36 pairs of values, i.e., $S = \{(6,6), (6, 5), \ldots (1,1)\}$. Because the dice are different colors, we consider outcomes as ordered pairs. As a result, outcomes $(5,6)$ and $(6,5)$ are different. 


### Countably Infinite Sample Space  

***Example: The Drowsy Programmer*** Program code written by a student is compiled. Let $C$ represent the outcome that the program compiles successfully and let $E$ represent the outcome that the compiler encounters an error. Suppose an experiment consists of selecting and compiling programs until a program is encountered that compiles on the first run.  
1. Define the sample space.  
**Solution:** The sample space is a countably infinite set $S = \{C, EC, EEC, EEEC, ...\}$  

2. What is the event that exactly one, three, or five programs are examined?  
**Solution:** Let A  be the event that exactly one, three, or five programs are examined. Then $A = \{C, EEC, EEEEC\}$.

Recall that we are assunming that there is a countably infinite number of program code files.  We could also state the problem in which there is a single program code which is compiled. If it fails, the programmer makes changes to the code  and then compiles again. This is repeated until the program sucessfully compiles.  Theretically, this can go on forever and hence countably infinite. Practically, the programmer should retake the appropriate programming language course if it goes on for too long!!


### Uncountably Infinite Sample Space

***Example: Stick Cutting*** Suppose that two points ($x_1$ and $x_2$) are marked (in some nondeterministic way) on a stick of length 1 meter.  

1. Define the sample space for this experiment.  
Since the sample space is portion of the two dimensional real plane, we know it is uncountably infinite. This will be the case for just about any infinitely variable real quantity.  
**Solution:** Let $x_1$ and $x_2$ denote the two points. Since $x_1$ and $x_2$ are marked randomly, sample space $S$ is given by
\begin{equation*}
       S ~=~ \{x_1 , x_2: 0 \leq x_1 \leq 1, 0 \leq x_2 \leq 1\}
\end{equation*}


2. Identify the event, "The second point is to the left of the first point."  
**Solution:** Let $G$ denote the  event: "The second point is to the left of the first point." Then the event $G$ is given by

\begin{equation*}
        G ~=~ \{x_1 , x_2 \in S ~s.t.~ x_1 > x_2\}
\end{equation*}

3. Suppose that the stick is cut at the marked points. Identify the event, "A triangle can be formed with the resulting three pieces."  
**Solution:** Let $Z$ denote the event that  "A triangle can be formed with the resulting three pieces." The first thing to note is that in any  triangle the sum of any two sides must be greater that the third side. For our stick problem this implies that none of the three pieces should be greater than  0.5 meters. There can be two cases: 1) $x_1 > x_2$ and 2) $x_1 \leq x_2$. 

Let's consider one case (intuitively, the other case will be symmetrical). If $x_1 > x_2$ then the three pieces will be of lengths $x_2$, $x_1 - x_2$, and $1 - x_1$. Due to the constraints each of these lengths must be less than 0.5. Thus, if we define 3 events A, B, and C

\begin{eqnarray*}
        A &:& x_2 < 0.5 \\
        B &:& x_1 - x_2 < 0.5 \\
        C &:& 1 - x_1 < 0.5
\end{eqnarray*}

Then, the event that the three pieces will form a triangle is $A \cap B \cap C$.  This is shown by the shaded area below the 45 degree line ($x_1 = x_2$) in Figure \@ref(fig:stick-cutting-problem). 


![Sample-Space](./Figures/stick-cutting-problem.svg)



Note that the case when $x_1 \leq x_2$ is symmetrical to the case $x_1 > x_2$. We can define events D, E and F which defined as follows

\begin{eqnarray*}
        D &:& x_1 < 0.5 \\
        E &:& x_2 - x_1 < 0.5 \\
        F &:& 1 - x_2 < 0.5
\end{eqnarray*}

and for this case, the shaded area above the 45 degree line in  Figure \@ref(fig:stick-cutting-problem) shows the set of points for which the triangle can be formed.  
In terms of the algebra of events, if we let $Z$ denotes the set of points that can form a triangle, then 

\begin{equation*}
        Z = (A \cap B \cap C) \cup (D \cap E \cap F)
\end{equation*}

## Naïve Definition Of Probability

If $A$ is an event in  the sample space $S$, then following is the naïve definition for the probability of event A, denoted by $P(A)$,

\begin{equation}
  P(A) = \frac{\text{Number of outcomes favorable to event A}}{\text{Total number of outcomes}} \tag{1-1}
\end{equation}


***Example: Double Coin Toss***  
1. What is the sample space $S$ for flipping a coin twice?  
**Solution:**
\begin{equation*}
S = \{ (H,H), (H,T), (T,H), (T,T)\}
\end{equation*}

2. What is the probability of the event $A = \text{Both Tails}$ using the naïve definition of probability?  
**Solution:**
\begin{equation*}
P(A) = \frac{1}{4}
\end{equation*}


There are two significant limitations of this naïve definition for general use of calculating probability of events. The naïve definition of probability is only applicable for  
1. finite sample space and  
2. equally likely outcomes.  
Despite the limitations, many problems can be solved using this naïve definition. And do not hesitate to use this whereever applicable. 



## Birthday Paradox

What is the probability that in a group of $k$ people at least two people share the same birthday?  

***Assumptions: ***  

1. There are 365 days, i.e., we are not considering leap year.  
2. The birthday of a person is equally likely to be any of the 365 days. Data shows that this not really true.  
3. We assume that the births are independent. The birth of one individual does not influence the birth of another person. Again this is not really true if you consider twins.  


For $k > 365$, $P(\text{atleast 2 people will have the same birthday}) ~=~ 1$. This is based on the Pigeon-hole principle.

For $k \leq 365$

\begin{equation*}
P(\text{no match}) ~=~ \frac{365 \times 364 \times \ldots \times (365 -k + 1)}{365^k}
\end{equation*}

What is the denominator? It is the number of possible ways of  birthdays of $k$ people. Infact this is the size of the sample space. 

What is the numerator? The number of ways they have different birthdays. 


\begin{eqnarray*}
P(\text{atleast 2 people with same birthday}) & = & 1 - P(\text{no match}) \\
& = & 1 - \frac{365 \times 364 \times \ldots \times (365 -k + 1)}{365^k} \\
& = & 50.7 \% ~~~~~\text{for k = 23} \\
& = & 97.1 \% ~~~~~\text{for k = 50} \\
& = & 99.99 \% ~~~~~\text{for k = 100}\\
\end{eqnarray*}


```r
n = 365
m = 100
prob = numeric(m)
for (k in 1:m) {
 num = 1
 for (i in 1:k) {
    num = num*(365-i+1) 
 }
 pnomatch = num/(365^k)
 prob[k] = 1 - pnomatch
}
plot(prob, ylab="Probability of match", xlab="Number of people (k)")
```

***Notes:***

1. Most people are surprised that with only 23 people there is a 50% chance  that two people will have the same birthday. The surprise is because 23 seems like a small number compared to 365.  
2. One way to reconcile is that with 23 people there are $\binom{23}{2} = 254$ different associations  being tested to see if there is match. This is a large number.  
3. As we will see above, the probability of a match as a function of $k$ is not a linear function.  
4. It is important to note that we have  found the probability that there is at least 1 match. It includes 1 match, 2 matches, 3 matches, and  so on. It is not so easy to find probability of exact number of matches, say, exactly 3 matches.  
5. It is important to clarify what we mean when we say 2 matches. Lets take an  example:  

    1. If A and B have the same birthday and C and D have a different same birthday, then we have two matches. No confusion here.

    2. If A, B, and C have the same birthday, is it 2 matches or 3 matches (A and B match, A and C match, B and C match)? This is considered to be 2 matches. We are looking for unique coincidences. If A and B match and B and C match then the match A and C is not longer a coincidence.

# Multiplication Rule
Consider a sequence of $r$ experiments $E_1, E_2, \ldots, E_r$. If experiment $E_1$ has $n_1$ possible outcomes and for each outcome of $E_1$, there are $n_2$ outcomes of $E_2$, and for each outcome of $E_2$, there are $n_3$ outcomes of $E_3$, $\ldots$, for each out of $E_{r-1}$ there are $n_r$ outcomes of $E_r$, then there are $n_1 \times n_2 \times \ldots \times n_r$ total possible outcomes. 

***Example: Choosing Coffee*** Customers of a café can customize their coffee in a number of different ways by selecting the type of coffee and the type of milk. We are considering only two features. In fact there are many more. Suppose there are two types of coffee ${\text{caffinated (caf)}, \text{decaffinated (decaf)}}$ and three types milk $\text{1 percent}, \text{2 percent}, \text{soy}$. The total number of outcomes (types of coffee) is $2 \times  3 = 6$. In writing so  we consider that the customer first chooses the type of coffee $E_1$ and then the type of milk $E_2$. This is shown in The solution would still be the same if we reversed the order. 


![Multiplication-Rule](./Figures/multiplication-rule.jpg)

*The number of ways in which a coffee can be ordered grows multiplicatively with each experiment*


From Figure \@ref(fig:multiplication-rule), we can see that as the number of experiments increase,  growth in the number of outcomes is very fast (exponential).

## Multinomial Coeffcients

A set of $n$ distinct objects is to be divided into $r$ distinct groups of sizes $n_1, n_2, \ldots n_r$ such that $\sum_{i=1}^r n_i = n$. The total number of ways $T$ is given by 


\begin{eqnarray}
T &=& \underbrace{\binom{n}{n_1}}_{\text{no. ways of choosing 1st group}} \underbrace{\binom{n - n_1}{n_2}}_{\text{no. ways of choosing 2nd group}} \ldots  \underbrace{\binom{n - n_1 - \ldots - n_{r-1}}{n_r}}_{\text{no. ways of choosing rth group}} \\
  &=& \frac{n!}{{n_1}! {n_2}! \ldots {n_r}!} 
\end{eqnarray}


***Example: Bridge*** In the game of bridge, how many ways can 52 cards be divided into 4 equal parts?   

**Solution:** Using the multinomial theorem we have:  
\begin{equation*} 
T = \frac{52!}{13! 13! 13! 13!}
\end{equation*} 

***Example: Choosing Library Pages*** The library employs 10 undergraduate students for different functions like shelving books, managing the checkout desk, help desk. The library has decide to have 5 students shelving books, 3 students working fulltime at the book checkout desk, and 2 students at the help desk. How many divisions ($T$) of the 10 students into the three possible groups are possible.   

**Solution:** We can apply the multinomial coefficients 
\begin{eqnarray*}
T &=& \frac{10!}{5! 3! 2!} \\
  &=& 252
\end{eqnarray*}

#### Notes 

1. Consider $( x_1 + x_2 + \ldots + x_r)^n$. The multinomial theorem states that 
\begin{equation*} 
( x_1 + x_2 + \ldots + x_r)^n =  \\
                \sum_{(k_1, k_2, \ldots, k_r): n_1 + n_2 + \ldots + n_r } \binom{n}{n_1, n_2, \ldots, n_r} x^{n_1}_1 x^{n_2}_1 \ldots x^{n_r}_1
\end{equation*} 
The sum is taken  over all possible values of $(n_1, n_2, \ldots, n_r)$ such that their sum is equal to $n$. The coeficient of the different terms of the expression are the multinomial coefficients.  
2. If we consider 2 groups with size  $n_1, n_2$, then $n_2 = n - n_1$. Then the total number of ways $T$ 

\begin{eqnarray*}
T &=& \frac{n!}{{n_1}! {n_2}!} \\
  &=& \frac{n!}{{n_1}! {n - n_!}!} \\
  &=& \binom{n}{n_1}
\end{eqnarray*}

which are the coefficients of the  Binomial Theorem. 

***Example: Arranging Balls***   
How many linear arrangements ($T$) are possible of $n$ balls of which $r$ balls are black and $n-r$ balls are white?   
**Solution:** There are $n$ positions. If we choose $r$ of those positions to place black balls, the remaining $n-r$ positions will have white balls. Thus

\end{equation*}


## Sampling

Suppose there are $n$ distinguishable objects. We want to find the number of ways we can pick $k$ objects out of $n$. We can consider two addtional features:   
a)  whether order matters or not and  
b)  with or without replacement.  

Let us first understand these two features

When we say *Order Does Not Matter*, what  this means is that different permutations of the same $k$  objects are not  considered to  be different outcomes. For example, ABC, ACB, BAC, BCA, BAC, CAB, CBA  are  not consider to be different outcomes. Whereas if Order Matters they are differnt outcomes. As an aside, we know that if  there are $k$ distinguishable objects, there are $k!$ different permutations/arrangements. 


When we say pick *Without Replacement*, what we mean is that when an object is chosen it is put aside and the next object is chosen from the remaining objects. On the other hand, when we replace the object before picking again,  we sample With Replacement.


Given the above two features, we  have the following 4 cases and the following table summarizes the results. 



|                      | Order matters            | Order does not matter    |
|----------------------|--------------------------|--------------------------|
| With Replacement     |  $n^k$                   |  ${n+k-1} \choose {k}$   |
|----------------------|--------------------------|--------------------------|
| Without Replacement  | $n(n-1) \ldots (n-k+1)$  |   $n \choose k$          |
|----------------------|--------------------------|--------------------------|




#### Case 1:  Order matters, With replacement

In this  case the first object  can be chosen in $n$ ways and since we are replacing  the object, the second object can also be chosen in $n$ ways. Using the multiplication rule we have a total of $n^k$ outcomes.

**Solution:**

2. The new system would generate $(3000^3)$ or 27 billion identifers, so the new system would be more than adequate.

## Inclusion-Exclusion Principle


First let us consider three events $A_1, A_2$, and $A_3$ as shown in  the figure below. 


![Three Events](./Figures/three-events-A-B-C.jpg)

*Three events A1, A2, and A3. The different intersections are shown. Note the area which is the intersection of all three events A1, A2, and A3.*

\begin{eqnarray}
P(A_1 \cup A_2 \cup A_3) &=&  P(A_1) + P(A_2) + P(A_3) \\
						 & &  - P(A_1 \cap A_2) - P(A_1 \cap A_3) - P(A_2 \cap A_3) \\
                         & &  + P(A_1 \cap A_2 \cap A_3)
\end{eqnarray}

1. When we add (include) $P(A_1) + P(A_2) + P(A_3)$ we have added the pairwise intersection area once too many. So in the second line we subtract (exclude). But in doing so we subtracted $A \cap B \cap C$ out completely. So we add (include) $A \cap B \cap C$. Due to the aternating addition and subtraction, the process is called the inclusion-exclusion principle.  

2. Collectively, the terms  in  the second line can be written as $\sum_{i < j}P(A_i \cap A_j)$ for $i \in \{1, 2, 3\}$ and $j \in \{1, 2, 3\}$.  

3. Also, note that the cardinality of the set $\{A_i \cap A_j\} \text{such that} ~i < j$ is $\binom{3}{2}$. 

4. Note that $\{A_i \cap A_j\}$ and $\{A_i, A_j\}$ mean the same thing and is used interchangeably. 



Let's consider 4 events $A_1, A_2, A_3, A_4$. 

\begin{eqnarray}
P(A_1 \cup A_2 \cup A_3 \cup A_4) &=&  P(A_1) + P(A_2) + P(A_3) + P(A_4) \\
						 & &  - P(A_1 \cap A_2) - P(A_1 \cap A_3) - P(A_1 \cap A_4) - P(A_2 \cap A_3) - P(A_2 \cap A_4) - P(A_3 \cap A_4)\\
                         & &  + P(A_1 \cap A_2 \cap A_3) + P(A_1 \cap A_2 \cap A_4) + P(A_1 \cap A_3 \cap A_4) + P(A_2 \cap A_3 \cap A_4) \\
                         & &  - P(A_1 \cap A_2 \cap A_3 \cap A_4)
\end{eqnarray}




Now we consider  the general case. If ${A_1}, {A_2}, \ldots, {A_n}$ are $n$ events, then

\begin{eqnarray}
  P(A_1 \cup A_2 \cup A_3 \cup \ldots \cup A_n) &=& \sum_{i=1}^{i=n} P(A_i)  \\
                                                & & - \sum_{{i_1} < {i_2}} P(A_{i_1} A_{i_2})\\
                                                & & +(-1)^{r+1}\sum_{{i_1} < {i_2} < ... < {i_r}} P(A_{i_1} A_{i_2}...A_{i_r})  + \ldots   ~~~~~~~~~ r = 3, \ldots, n-1 \\
                                                & & +(-1)^{n+1} P(A_1 A_2 A_3 \ldots A_n)
\end{eqnarray}

where ${i_1} < {i_2} < ... < {i_r}$ refers to all possible r-wise combinations of $(1, 2, \ldots , r)$. 


What the above equation states is that the probability of the union of $n$ events is equal to the sum of the probabilities of the of events taken one at a time, minus the probability of the intersection of events taken two at a time, plus the probability of the events taken three at a time, and so on until the probability of the intersection of events all taken together.


## Proof of the Inclusion-Exclusion Principle

The inclusion-exclusion identity can be proved by induction. Below we show one step of the induction. In partiucular,  assuming that the equation is true for $n = 2$, we derive the expression for $n = 3$. That is derive the expression for
$P(A_1 \cup A_2 \cup A_3)$  given the $P(A_1 \cup A_2) = P(A_1) + P(A_2) - P(A_1 \cap A_2)$. In doing this step we will convince ourselves that  the above general Equation \@ref(eq:inclusion-exclusion-g) is true. 

We are given that for any two events $E_1$ and $E_2$, $P(E_1 \cup E_2) = P(E_1) + P(E_2) - P(E_1 E_2)$.

Now consider three events $E_1$, $E_2$, and $E_3$. $E_1 \cup E_2 \cup E_3$ can be written as $E_{12} \cup E_3$ where $E_{12}$ is $E_1 \cup E_2$.
\begin{eqnarray}
P(E_1 \cup E_2 \cup E_3) &=& P(E_{12} \cup E_3) \\
                            &=& P(E_{12}) + P(E_3) - P(E_{12} E_3) \\
                            &=& P(E_1) + P(E_2) - P(E_1 E_2) + P(E_3) - P(E_{12} E_3)
\end{eqnarray}

Now, $E_{12} E_3 = (E_1 \cup E_2)\cap E_3$ which using the distributive law is $(E_1 \cap E_3) \cup (E_2 \cap E_3)$. Now we can use the inclusion-exclusion identity for two events to obtain the following

\begin{eqnarray}
P(E_{12} E_3) &=& P((E_1 \cap E_3) \cup (E_2 \cap E_3))\\
                &=& P(E_1 E_3) + P(E_2 E_3) - P(E_1 E_3 E_2 E_3) \\ 
                &=& P(E_1 E_3) + P(E_2 E_3) - P(E_1 E_3 E_2)
\end{eqnarray}

The last step is because $E_3 \cap E_3 = E_3$.  Now replace $P(E_{12} E_3)$ in Equation \@ref(eq:inclusion-exclusion-3i) to obtain the final result.


## Example: Network Recieve Buffer
A server has 10 incoming packet flows (flows are distinct and uniquely identified by a 5-tuple flow-id). Packets from each flow are stored in a buffer. Suppose there are two packets from each flow (packets have sequence numbers and hence are distinguishable) which are randomly assigned to locations in the buffer of size 20. What is the probability that no two packets from the same flow are in adjacent locations in the buffer?   

![Network-IO-Buffer](./Figures/network-io-buffer.jpg)

**Solution:** 

Let $E_i, i= 1, 2, 3, \ldots, 10$, be the event that packets from the ith flow are placed in adjacent locations in  the buffer. Then the required probability $\alpha$ is given by 
\begin{equation*}
\alpha = 1 - P(\bigcup_{i=1}^{10} E_i)
\end{equation*}

Why? 

What we want is  $P(\bar{E_1} \bigcap \bar{E_2} \ldots \bigcap \bar{E_{10}})$. This by de Morgan's law is $P(\overline{{E_1} \bigcup {E_2}  \ldots \bigcup {E_{10}}})$. We then use the first property of probability. 

Now reflect on the fact that the events are not non-overlapping. Hence, using the Inclusion-Exclusion principle, we have 
\begin{equation*}
P(\bigcup_{i=1}^{10} E_i) = \sum_1^{10} P(E_i) + \ldots + (-1)^n \sum_{i_1 < i_2 < \ldots < i_n} P(E_{i_1} E_{i_2}  \ldots E_{i_n}) 
+ \ldots - P(E_1 E_2 \ldots E_{10})
\end{equation*}
Now we have to find the probabilities, i.e., $P(E_{i_1} E_{i_2}  \ldots  E_{i_n})$ which is the probability that packets from $n$ flows are placed adjacent to each other. Now the total number of ways in which 20 packets can be arranged in the buffer locations is $20!$ and these are equally likely outcomes. If we consider the packets from the flows that are adjacent as single elements, then the total number of possible orderings is $(20 - n)!$. Thus, 

\begin{equation*}
P(E_{i_1} E_{i_2}  \ldots E_{i_n})  = \binom{10}{n} 2^n \frac{(20 -n)!}{20!}
\end{equation*}

The first term in the RHS is the number of ways we can choose the $n$ flows. The second term comes from the fact that there are two ways in arranging the two packets in each flow that are together.  Thus we get

\begin{eqnarray}
P(\bigcup_{i=1}^{10} E_i) &=&  \binom{10}{1} 2^1 \frac{19!}{20!} -  \binom{10}{2} 2^2 \frac{18!}{20!} +  \binom{10}{3} 2^3 \frac{17!}{20!} - \ldots - \binom{10}{10} 2^{10} \frac{10!}{20!} \\
                          &\approx& 0.6416608
\end{eqnarray}

Let's write the code to compute it. 

```r
sum = 0  
n = 10
for (i in 1:n)
{
  sum = sum + (-1)^(i+1)*(factorial(n)/(factorial(n - i)*factorial(i)))*(2^i)*factorial(2*n - i)/factorial(2*n)
}
print(sum)
```

Thus, $\alpha = 1 - 0.6416608 =  0.3583392$

# Matching Problem

There is a deck of $n$ cards labelled $1, \ldots, n$. The  cards are shuffled. Call out numbers 1 through $n$ in sequence and for each call flip a card from the deck. What is the probability that for at least one called number, the corresponding card number will match? This problem was first considered by Pierre-Remond  Montemort's problem around 1713.  

Another version of the problem is called the Hat matching problem.
Suppose $n$ people throw their hats in a room. Then they randomly pick a hat. What is the probability that at least one person picks his own hat.


Let $A_j$  be the event that the $j$th card matches.

\begin{eqnarray}
P(\text{at least 1 match}) &=& P(A_1 \cup A_2 \cup A_3 \cup \ldots \cup  A_n) \\
              &=& \sum_{i=1}^{i=n} P(A_i)  \\
              & & - \sum_{{i_1} < {i_2}} P(A_{i_1} \cap A_{i_2})\\
              & & +  (-1)^{r+1}\sum_{{i_1} < {i_2} < ... < {i_r}} P(A_{i_1} \cap A_{i_2} \cap \ldots \cap A_{i_r}) + \ldots \\
              & & + (-1)^{n+1} P(A_1 \cap A_2 \cap A_3 \cap \ldots \cap A_n)
\end{eqnarray}

The above is based on the Inclusion-Exclusion principle.  We will now determine each of those terms. 

1. The single terms 
\begin{equation}
P(A_j) = \frac{1}{n}  ~~~~~ \text{for all $j = 1, \ldots, n$}
\end{equation}
This comes from the observation that all positions are equally likely for card labelled $j$.  
How many tems are there? $n$ which we could also write as $n \choose 1$. 

2. Terms taken two events at a time
\begin{equation}
P(A_i \cap A_j) = \frac{(n-2)!}{n!}  ~~~~~ \text{for all} ~~~ i, j, = 1, \ldots, n ~~~~\text{and}~~~~ i > j
\end{equation}
This is based on the observation that 2 cards $A_i$ and $A_j$ are in the right positions  so the remaining $n - 2$ can be permuted in $(n - 2)!$ ways which gives the total number of favorable outcomes out of $n!$ which is the total number of outcomes.  
How many tems are there? There are  $n \choose 2$ terms. 

3. We can generalize this to $k$ events taken together 
\begin{equation}
P(A_{i1} \cap A_{i2} \ldots \cap A_{ik}) = \frac{(n-k)!}{n!}  ~~~~~ \text{for all}~~~ i_1, i_2, \ldots i_k = 1, \ldots, n ~~~~\text{and}~~~i_1 > i_2 > \ldots > i_k
\end{equation}
How many tems are there? There are  $n \choose k$ terms. 


4. 
Now we put it all together

\begin{eqnarray*}
P(\text{at least 1 match}) &=& \binom{n}{1} \times \frac{1}{n}  \\
                           & & - \binom{n}{2} \times \frac{(n-2)!}{n!}  \\
                           & & +(-1)^{k+1} \binom{n}{k} \times \frac{(n-k)!}{n!}   ~~~~ k = 3, \ldots, n-1 \\
                           & & +(-1)^{n+1} \binom{n}{n} \times \frac{(n-n)!}{n!} \\
                           &= & \frac{1}{1!} - \frac{1}{2!} + \frac{1}{3!} + \ldots + (-1)^{n+1} \frac{1}{n!} \\
                          &\approx& 1 - \frac{1}{e}
\end{eqnarray*}

Couple of notes:

1. To get the final result we use the asymptotic value of the Taylor series as $n \rightarrow \infty$.  
2. Why is there an asymptotic value? As  $n \rightarrow \infty$, the chance of getting a match becomes smaller and smaller however, the number of attempts for a match becomes larger and larger. These two opposite forces converges to $1 - \frac{1}{e}$.

\begin{eqnarray*}

\end{eqnarray*}

# Conditional Probability

We started with the naïve definition of probability and solved some challenging and important problems. We discussed the limitations of the naïve definition and then gave a formal definiton of probability. Now we take the next step to increase the power of probabilistic modeling. In particular, we discuss how we update  the evaluation of the uncertainty (risk) when we are given new data. In a sense this is what we as humans are constantly doing--re-evaluating risk based on data that we sense from the environment.  Conditional probability is also considered the foundation for statistics. 

***Definition of Conditional Probability***

\begin{equation*}
    P(A|B) = \frac{P(A \cap B)}{P(B)}  ~~~~~~~~~~~ P(B) > 0
\end{equation*}

In Figure \@ref(fig:cond) given that event $B$ has occurred the relevant portion of event $A$ is $A \cap B$. Given that $B$ has occurred, outcomes that are not in $B$ ($B^{\mathsf(c)}$)are no longer relevant. The sample space has now collapsed to $B$ since only outcomes in $B$ are possible. 


![Conditional Probability](Figures/Conditional-probability.jpg)



### Few Theorems

1. Theorem 1  
\begin{eqnarray}
  P(A \cap B) &=& P(B) \times P(A|B) \\
  			  &=& P(A) \times P(B|A)
\end{eqnarray}

Note that $P(A \cap B) = P(B \cap A)$

2. Multiplication Rule  
\begin{equation*}
    P(A\cap B \cap C) = P(A) \times P(B|A) \times P(C|A \cap B)
\end{equation*}


3. Bayes' Rule  
\begin{equation*}
   P(A|B) =  \frac{P(A) \times P(B|A)}{P(B)}
\end{equation*}  
$P(A|B)$ is called the posterior probability of A and $P(A)$ is called the prior. Note that $P(B)$ can also be written as  
\begin{equation}
  P(B) = P(A) \times P(B|A) + P(A^\mathsf{c}) \times P(B|A^\mathsf{c})
\end{equation}

4. The Law of Total Probability: If $A_1, A_2, \ldots A_n$ are mutually exclusive and exhaustive events, then

\begin{equation*}
        P(B) = P(B|A_1)\times P(A_1) + P(B|A_2)\times P(A_2) + \ldots + P(B|A_n)\times P(A_n)
\end{equation*}


The above theorems can be  proved using the axiom of probability and the definition of conditional probability.

***Example: Testing for a Rare Disease***  
Patients get tested for a disease that affects 1\% of the population. The test is 95\% accurate. If a patient tests positive what is the probability that the patient has the disease?  
A common (wrong) answer is 95\%.  
Let  
1. ***T***: event that a patient tests positive  
2. ***D***: event  that a patient has the disease  

We will assume that the test is 95\% accurate implies that

\begin{eqnarray*}
P(T|D) &=& 0.95 \\
P(T^\mathsf{c} |D) &=& 0.05 \\
P(T^\mathsf{c} |D^\mathsf{c}) &=& 0.95 \\
P(T |D^\mathsf{c}) &=& 0.05
\end{eqnarray*}

Let's make sure we understand the above terms. The terms are a formal re-statement of the properties of the test. Recall that "95\%" accurate implies that the test is 95\% accurate in both directions--for testing the presence or absense of the disease. Hence, in this scenario, 

1. $P(T|D)$ can be thought of as the "true positive" conditional probability. For example, suppose that the person tested positive for the disease and then later presented symptoms in a clinical setting. 

2. $P(T^\mathsf{c} |D)$ is the "false negative conditional probability. Suppose in this case that the person tested negative for the disease, but then presented symptoms later in a clinical setting. 

3. $P(T^\mathsf{c} |D^\mathsf{c})$ is the "true negative" conditional probability--the odds that someone tested negative and does not, in fact, have the disease. 

4. $P(T |D^\mathsf{c})$ is the "false positive" conditional probability. This patient tested positive for the disease, but never presented symptoms.  

We now use Bayes' Theorem  
\begin{eqnarray*}
P(D|T) &=& \frac{P(D) \times P(T|D)}{P(T)} \\
       &=& \frac{P(D) \times P(T|D)}{P(D) \times P(T|D) + (D^\mathsf{c}) \times P(T|D^\mathsf{c}) } \\
        &=& \frac{0.01 \times 0.95}{0.95 \times 0.01 + 0.05 \times 0.99 } \\
       &=& 0.16
\end{eqnarray*}
Here is another way of arriving at the solution: Let say we have a population of 10,000. Since 1\% of the population has the disease, this implies that about 100  people will have the disease and 9900 people will not have the disease. Even though the test is reasonably accurate, 5\% of 9900 which is 495 people who do not have the disease will test positive. And 95 of the hundred people who have the disease will test positive. Now if a person test positive, the probability that she has the disease is $\frac{95}{95+495}$ which is 16\%. This shown in the following figure 

![Test-for-disease](Figures/Test-for-disease.jpg)


The important point to note here is that even though the test is reasonably accurate, a significant number who do not have the disease will test positive. You can also think of this as a "signal to noise" problem. The "signal" is the fact that this only affects 1\% of the population, but the noise is 5\% of inaccuracy of the test. 

This contradicats the common fallacy that $P(A|B) = P(B|A)$ with the introduction of new information. Notice how the fact that the disease only affects 1\% of the population was subtly conveyed, but indeed was key to calculating $P(D)$.

### The Law of Total Probability

In this case we use conditional probability to determine the probability $P(B)$ of an event B. If $A_1, A_2, \ldots A_n$ are mutually exclusive and exhaustive events, then

\begin{eqnarray*}
        P(B) &=& P(B \cap A_1) \cup P(B \cap A_2) \cup \ldots \cup P(B \cap A_n) \\
            &=& P(B \cap A_1) + P(B \cap A_2) + \ldots \cup P(B \cap A_n) \\
            &=& P(B|A_1)\times P(A_1) + P(B|A_2)\times P(A_2) + \ldots + P(B|A_n)\times P(A_n)
\end{eqnarray*}

This is best illustrated in Figure \@ref(fig:Lawoftotalprobability).


![Law-of-total-probability](Figures/Lawoftotalprobability.jpg)



It is important to note  that $A_1, A_2, \ldots A_n$ are mutually exclusive and exhaustive events.  Note that a simple case is that we  can use $A$ and $A^\complement$.

\begin{eqnarray*}
        P(B) &=& P(B \cap A) \cup P(B \cap A^\complement)\\
            &=& P(B \cap A) + P(B \cap A^\complement) \\
            &=& P(B|A)\times P(A) + P(B|A^\complement)\times P(A^\complement)
\end{eqnarray*}


***Example: A Conditional Urn Problem***

This following problem is difficult and shows the use of Law of Total Probability  and Bayes' Theorem. As you go through it, remember that $P(\ast|F)$ is complete probability space and all axioms of probability applies. Only the sample space is now changed.


There is a one ball in a bag. It is equally likely that the ball is either red or blue. A red ball is now added to the bag and then a random ball is taken out. If the ball withdrawn is red what is the probability that the remaining ball is also red?

We define the following events

1. A: event that the initial ball is red
2. B: event that  the removed ball is red
3. C: event that the remaining ball is red


We need to find $P(C|B)$. To find this by conditioning whether the initial marble is red.

\begin{eqnarray*}
  P(C|B) &=& P(C|B \cap A) P(A|B) + P(C|B \cap A^\complement) P(A^\complement |B) \\
         &=& 1 \times  P(A|B) + 0 \times   P(A^\complement|B)\\
         &=& P(A|B)
\end{eqnarray*}


To find $P(A|B)$ we use Bayes' Rule

\begin{eqnarray*}
  P(A|B) &=& \frac{P(B|A) P(A)}{P(B)} \\
         &=& \frac{\frac{1}{2}}{P(B|A) P(A) + P(B|A^\complement) P(A^\complement)} \\
         &=& \frac{\frac{1}{2}}{\frac{1}{2} + \frac{1}{4}} \\
         &=& \frac{2}{3}
\end{eqnarray*}


Thus the required probability $P(C|B) = 2/3$.



### Prosecutor's Fallacy
Police in a city of 1 million ($10^{6}$) are working on a crime. Clearly. since 1 in 10,000 fit the description, with $10^{6}$ people, there will be 100 people who will fit the description. So the probability that the defendant is guilty is 1 in 100 which is small. We want to clearly understand where the prosecutor is making wrong statement.

Let  $D$ denote the event the person fits the description and let $I$ denote the event that a person is innocent. 

Clearly, S1 which restate the statistics that 1 in 10,000 people fit the description is true.

       &=& \frac{99}{999,999} \\
       &=& 9.9 \times 10^{-5}
\end{eqnarray*}

This is indeed small and hence statement S2 is true.

\begin{eqnarray*}
P(I|D) &=& \frac{P(D \cap I)}{P(D)} \\
       &=& \frac{99}{100} \\
       &=& 0.99
\end{eqnarray*}

which is very close to 1.

Hence, statement S3 is wrong. The prosecutor in essence is wrongly equating $P(I|D)$ to be the same as $P(D|I)$ but they are not.

## Independence of Events 

We  consider two events $A$ and $B$ to be independent if 

\begin{equation}
P(A \cap B) = P(A) \times P(B) 
\end{equation} 


What this intuitively means is that that the occurence of Event A does not affect the occurrence of Event B. It is important to note that this is different from Events A and B being mutually exclusive. Infact, if two events are mutually exclusive they are not independent. 


***Example: Two Randomly-Drawn Cards***  
Suppose a card is drawn at random from a deck of cards. Let A be the event that the card is a spade and B be the event that card is an Ace. These two events are independent. $P(A) = 13/52 = 1/4$ and $P(B) = 4/52 = 1/13$ and $P(A \cap B) = 1/52 = P(A) \times P(B)$


***Example: Odds of Having a Baby Boy or Girl***

## Variance of a Discrete Random Variable

If $X$ is a discrete random variable and takes values $x_1, x_2, x_3, \ldots , x_n$  then the variance of $X$ denoted by $\text{Var}[X]$ is  given by 

\begin{eqnarray}
Var(X) &=&  E(X^2) - (E(X))^2\\
       &=& \sum_{i=1}^{n} p_i x_i^2 - (\sum_{i=1}^{n} p_i x_i)^2
\end{eqnarray}

where $p_i = P(X = x_i) $. 

1. $E(X^2)$ is called the second moment. 
2. The standard deviation typically denoted by $\sigma$ is the square root of the variance, i.e, $\sigma = \sqrt{Var(X)}$. 
3. The standard deviation measures how the probability mass is distributed around the mean. The large the value of $\sigma$ the wider the spread. 


### Expectation and Variance of Bernoulli

\begin{equation}
E(X) = 1 \cdot p + 0 \cdot (1 - p) = p
\end{equation}  

\begin{eqnarray}
Var(X) &=& E(X^2) - (E(X))^2 \\ 
       &=& p - p^2 \\
       &=&  p(1-p)
\end{eqnarray}

Suppose $X \sim \text{Bern}(0.6)$.  

\begin{equation}
X = 
\begin{cases}
	1 & \text{with probability $p = 0.6$} \\
	0 & \text{with probability $1-p = 0.4$}
\end{cases}
\end{equation}

## Binomial Distribution
We can see that the Bernoulli Random Variable is useful for some relatively simple experiments, but what if we want to know how many successes we can expect to have in a series of Bernoulli trials? For that, we have the Binomial Random Variable: It expresses  number of successes $X$ in $n$ independent $\text{Bern}(p)$ trials where $p$ is the probability of success. Its distribution is given by:  

\begin{equation}
        P(X=k) = {\binom{n}{k}} p^k (1-p)^{n-k}  ~~~~~~~~~ 0 \leq k \leq n
\end{equation}

Note that Binomial random variable has 2 parameters $n$ and $p$ and will be written as $X \sim \text{Binom}(n,p)$.

### Expectation and Variance

\begin{equation}
E(X) = np
\end{equation}  

\begin{equation}
Var(X) = np(1-p)
\end{equation}

***Example: Failure modes for a communications system***
A communication system consists of $n$ components, each of which will independently function with probability $p$. The total system will be able to operate effectively if at least one-half of its components function. For what value of $p$ is a 5-component system more likely to operate effectively than a 3-component system?

**Solution:**  
Let $X_5$ be the random variable that denotes number of functioning components in a 5-component system. $X_5$ has a Binomial distribution with parameters $(n=5, p)$. This is written as $X_5 \sim \text{Binom}(5,p)$. Similarly, we can define $X_3$ to be the random variable that denotes number of functioning components in a 3-component system. Again, $X_3 \sim \text{Binom}(3,p)$ The probability that the 5-component system is functioning effectively is given by $P(X_5 > 2)$ which is given by

\begin{eqnarray*}
  P(X_5 > 2)  & = & \underbrace{{\binom{5}{3}} p^3 (1-p)^2}_{P(\text{3 out of 5 functioning})} + \underbrace{{\binom{5}{4}} p^4 (1-p)^1}_{P(\text{4 out of 5 functioning})} + \underbrace{{\binom{5}{5}}) p^5 (1-p)^0}_{P(\text{all 5 functioning})} \\
              & = & {\binom{5}{3}} p^3 (1-p)^2 + {\binom{5}{4}} p^4 (1-p) + p^5
\end{eqnarray*}


Similarly, the probability that the 3-component system is functioning is given by

\begin{eqnarray*}
  P(X_3 > 1)  & = & {\binom{3}{2}} p^2 (1-p)^1 + {\binom{3}{3}} p^3 (1-p)^0 \\
              & = & {\binom{3}{2}} p^2 (1-p)^1 +  p^3
\end{eqnarray*}


For the 5-component system to  more likely  operate than a 3-component system we want $P(X_5 > 2) > P(X_3 > 1)$. From which solving for $p$ we can show that $p > 0.5$.

## Geometric Distribution 

We consider independent $\text{Bern} (p)$ trials. Let the random variable $X$ denote the number of trial until  first success. Then $X$ has a Geometric distribution which is written as $X \sim \text{Geom}(p)$. It turns out there are actually two different geometric distributions. In one, we want to know the number of trials **up to and including** the first success  The pmf is given by

\begin{equation}
	P(X=k) = (1 - p)^{k-1} p ~~~~~~~~ k  \in \{1, 2, 3, \ldots, \infty \}
\end{equation}

The CDF is given by
\begin{eqnarray*}
F_X (x) &=& P(X \leq x) \\
        &=& \sum_{i = 1}^{\lfloor{x}\rfloor} P(X = i)
\end{eqnarray*}

The Expectation is:  
\begin{equation}
\frac{1}{p}
\end{equation}

In the other kind of geometric distribution, we just want to know the number of failures **before** the first success. Formally, let the random variable $Y$ denote the number of failures until first success. Then $Y$ has a Geometric distribution which is written as $Y \sim \text{Geom}^\text{*} (p)$.  The pmf  is given by:

\begin{equation}
	P(Y=k) = (1 - p)^{k} p ~~~~~~~~ k \in \{0, 1, 2, 3, \ldots, \infty \}
\end{equation}

Note that in this case the range of the random variable starts from 0. 

The Expectation is:  
\begin{equation}
\frac{1-p}{p}
\end{equation}

The Variance for either of the above two Geometric distributions is:  
\begin{equation}
\frac{1-p}{p^2}
\end{equation}


### Properties

Caution! Here lies our first Random Variable which occupies a countably infinite sample space! Why? Because unlike the Binomial and Bernoulli, we are no longer limiting the number of trials. Bernoulli problems only concerned 1 trial and the Binomial was only concerned with how many times an outcome would occur with $n$ trials. So you'll see this extra "parameter" $k$ just floating around. In theory, $k$ is functioning like our little $x$ in the previous Random Variables. In practice, you can think of it as our "sanity" parameter; keeping us safely in the bounds of numbers that we care about and can actually compute. We will be varying k in practice over some finite vector of numbers.

1. The Geometric distribution has only one *true* parameter: $p$. 

2. The values that the random variable takes is countably infinite, i.e., the set of postive integers. 

3. Show that the above pmf is valid. This is done by showing two properties. First, we need to show that $P(X = i) \geq 0$ for all $i = 0, 1, 2, \ldots$. Second, we need to show that $\sum_{i=0}^{\infty} P(X = i) = 1$. These are easily shown.  

4. This distribution has a interesting property called the memoryless property. More about this after when we study the Poisson random variable.

## Medium Access Control (MAC) Protocol
Suppose $n$ nodes share a common broadcast channel. This is shown in Figure \@ref(fig:medium-access-control) What that means is that when one node transmits a packet it is heard by all the other nodes. Consequently, if two or more nodes transmit packets "near simultaneously", their transmissions will collide and the bits in the packet will be garbled.



![medium-access-control](Figures/MAC_Protocol.jpg)


A important goal is to design a distributed MAC protocol that minimizes the collisions. To study this we make the following assumptions:  
1. There are $n$ independent and identical nodes  
2. Time is divided into slots  
3. All transmission attempts are made at the beginning of a time slot  
4. If a packet arrives in a time slot, the station will attempt its transmission at the beginning of the next slot  
5. At each node, at most one packet can arrive in a timeslot and at each node at any given timeslot, the probability that a packet will arrive is denoted by $p$  
6. There is no buffer in the node. If a packet is not "successfully" transmitted it is  lost  

For any given slot, let $X$ denote the random variable  that a node has a packet to transmit. If we define Success to denote that the node has a packet to transmit, then $X \sim \text{Bern}(p)$.  Hence,

\begin{eqnarray*}
P(\text{success}) &=& P(X = 1) \\
                  &=& p \\
P(\text{failure}) &=& P(X = 0) \\
                  &=& 1 - p
\end{eqnarray*}

In any given time slot, let $Y$ be the random variable that denotes the number of nodes that have packets to send. Nodes that have packets to send will be referred to as active nodes. $Y$  an take vales $\{0, 1, 2, \ldots ,n\}$. Since the nodes are independent $Y \sim \text{Binom}(n,p)$. Hence we have  the PMF which gives the probability that $k$ nodes are active

\begin{eqnarray*}
P(Y=k) &=& P(\text{$k$ nodes are active})  \\
       &=& \binom{n}{k}p^k (1-p)^{n-k}
\end{eqnarray*}


In two or more node transmit in the same slot, then their transmissions will  collide and both the packets will be garbled. We will refer to this as a collision. Let $P_c$ denote  the probability of collision and this given by

\begin{eqnarray*}
P_c &=& P(\text{two or more nodes are active in a slot}) \\
    &=& \sum_{k=2}^{n} P(Y = k) \\
    &=& 1 - P(Y = 0) - P(Y = 1) \\
    &=& 1 - \binom{n}{0} p^0 (1-p)^n - \binom{n}{1} p^1 (1-p)^{n-1} \\
    &=& 1 - (1-p)^n - np(1-p)^{n-1}
\end{eqnarray*}

One of the goal of the MAC protocol is to minimize the number of collisions. Can you find the the value of $p$ that minimize $P_c$?  
We can also find the following two probabilities  
\begin{eqnarray*}
P_s &=& P(\text{successful transmission in a slot}) \\
    &=& P(Y = 1) \\
    &=& np(1-p)^{n-1}
\end{eqnarray*}

and

\begin{eqnarray*}
P_s &=& P(\text{slot is idle, i.e., no packet transmission attempt}) \\
    &=& P(Y = 0) \\
    &=& (1-p)^{n}
\end{eqnarray*}


One of the assumptions we made is that there is no buffer  in each node. What that means is that if there is a collision, packets in all the nodes that collide are lost. We make this assumption to simplify the model. In reality, there will be a buffer and the packet may be re-attempted in a subsequent slot. A consequence of our assumption is that we can treat the outcomes in the slots to be independent. In other words, what happens in a slot is not impacted by what happened in the the previous slots.

Let $Z$ be a random variable that denotes the number of unsuccessful (collision or idle) slots before a successful transmission. $Z$ can take values $0, 1, 2, \ldots $. If $S, I, C$ denote the successful transmission, idle, and collision, in a slot, respectively. Then, ICS would correspond to $Z = 2$. We can see that $Z \sim Geom(p)$ where $p = P_s$ and the pmf is given by

\begin{equation*}
P(Z = k) = (1 - p)^k p  ~~~~~~~~~~~ k \in \{0, 1, 2, \ldots \}
\end{equation*}

and so $P(Z=4) = (1-p)^4 p$.


## Poisson Process and the Poisson Distribution


![poisson-process](Figures/Poisson-Process.jpg)



Consider that events occur in time. Let $N(t)$ be a counting process that counts the number of events that occur in some interval of time $t \geq 0$. $N(t)$ is a Poisson process with rate $\lambda$ if the following properties are satisfied:

1. $N(0) = 0$;  
2. The number of events that occur in disjoint time intervals are independent;  
3. The distribution of the number of events in a given interval depends only on the length of the interval and not on the location;  
4. For small values of $h$,  
\begin{equation*}
P(N(h) =1) = \lambda h + o(h)
\end{equation*}
where $o(h)$ is any function $f(h)$ for which  
\begin{equation*}
lim_{h \rightarrow 0} \frac{f(h)}{h} =0
\end{equation*}  
5. $P(N(h) = 2) = o(h)$.  

Yes, it's true that this Random Variable goes to infinity, like the Geometric Random Variable. What do they have in common? They're both counting the number of times something happens with some parameters. Remember that with the Binomial Random Variable, we were simply asking the probability that out of a finite number of trials, we had "success" some finite number of times. The Bernoulli was even more limited to just the single trial.

If $N(t)  \sim \text{Pois} (\lambda)$ then the pmf is given by  

\begin{equation*}
P(N(t) = k) = \frac{e^{-\lambda t} (\lambda t)^k}{k!} ~~~~~~~~~~ k \in \{0, 1, 2, \ldots \}
\end{equation*}


***Example: Requests to a Web Server***

\begin{eqnarray*}
P(N(10)= 3)  &=&  \frac{e^{-\lambda 10} (\lambda 10)^3}{3!} \\
					 &=&  \frac{e^{-3} (3)^3}{3!}
\end{eqnarray*}

### Expectation and Variance of the Poisson Distribution
They're both just $\lambda$. For the expected value, this kind of makes sense intuitively when you recall that $\lambda$ is, after all the "rate" parameter. Broadly, both the Expectation and the Variance stem from the fact that:  
\begin{equation}
e^x=\sum_{k=0}^\infty \frac{x^k}{k!}
\end{equation}

by the Taylor Series Expansion of the Exponential.

### Relationship Between Poisson Distribution and Binomial Distribution


Events $A_i, A_2, A_3, \ldots, A_n$ are events with $P(A_j) =p_j$. If $n$ is large and $p_j$'s are small and if the events are independent or "weakly dependent," then the number if events that occur is approximately $\text{Pois} (\lambda)$ where $\lambda = \sum_{j=1}^n p_j$.

If the events are independent   and $p_j =p \forall j$, then we have Bernoulli trial with large $n$ and small $p$. The number of events that occurs $X$ is a random variable where $X \sim \textbf{Binom}(n. p)$. When $n \rightarrow \infty$ and $p \rightarrow 0$ such that $np$ is constant, then $X$ can be well approximated by $\text{Pois} (\lambda)$ where $\lambda = np$.

### Relationship Between Poisson Process and Exponential Distribution

Consider events that follow the Poisson process with parameter $\lambda$. Let $X$ be a random variable that denotes the waiting time between events; i.e., starting from an event what is the time until the next event. This is also called the inter-arrival time.

which means that $X \sim \text{Expo} (\lambda)$.

    E(T) &=& E(T_1 + T_2 + \ldots + T_n) \\
         &=& E(T_1) + E(T_2) + \ldots + E(T_n) \\
         &=& 1 + \frac{n}{n-1} + \frac{n}{n-2} + \ldots + \frac{n}{1}\\
         &=& n \underbrace{(1 + \frac{1}{2} + \frac{1}{3} + \ldots + \frac{1}{n})}_{\text{Harmonic Series}} \\
         &\approx& n ~log(n) + 0.5772~n + 0.5
\end{eqnarray*}

## Expectation

The mean/expectation/average value of a random variable $X$ is denoted as $E(X)$. It is simply the result of multiplying each possible value of $X$ by its probability and summing.
\begin{equation}
E(X) = \sum_{x \in X } x \times P(X=x)
\end{equation}
Note that $x \in X$ implies over all values that the random variable takes.

1. This is called the weighted average of the different values the random variable takes, where the weights are the probabilities.  
2. In the summation we need only consider non-zero values that the random variable takes.  
3. The mean value need not be a value that the random variable takes.  


***Example: Tossing coins***
Suppose three coins are tossed.  We can define the following random variable
\begin{equation*}
Y = \text{The number of heads that appear}
\end{equation*}
What is $E(Y)$?

We can show that the values $Y$ can take are $\{0, 1, 2, 3\}$ and the pmf is $P(Y=0) = 0.125, P(Y=1) = 0.375, P(Y=2) = 0.375, P(Y=3) = 0.125$. Now $E(Y)$ is given by
\begin{eqnarray*}
E(Y) &=& 0 \times 0.125 + 1 \times 0.375 + 2 \times 0.375 + 3  \times 0.125 \\
     &=& 0.375 + 0.750 + 0.125  \\
     &=& 1.25
\end{eqnarray*}

Notice that we only need to consider non-zero values that the random variable takes. The first term in LHS does not contribute anything. Also, notice that the mean value 1.25 is not really a value that the random variable takes. In that sense the mean value is kind of strange.


***Example: Roulette***  
Suppose you play roulette and bet 10 dollars on red. Assuming an American table, you would win $10$ dollars with probability 18/38 and lose $10$ dollars  with probability 20/38. Let $X$ be the random variable representing your winnings. Using the above formula, the expected value of your winnings would be 
\begin{equation}
E(X) = 10 \cdot \frac{18}{38} + (-10) \cdot \frac{20}{38} = -0.526
\end{equation}
or about ($0.53$ dollars) in accountant's notation. Clearly a player at a roulette table won't be seen shaving $0.53$ dollars off their $10$ dollars chip on each play or tossing \$0.53$ at the dealer while shouting something about expected value, but instead this makes use of the Law of Large Numbers, which we will investigate in detail later when considering Limit Theorems. In short, however, it can be surmised that over many plays, the player will lose roughly $0.53$ dollars  per $10$ dollars  bet with very high probability.

## Some Properties of Expectation

### Expectation of a Function of a Random Variable

Suppose $g(X)$ is some function of the random variable $X$.  We will state the following property without proving it.
\begin{equation}
E(g(X)) = \sum_{x \in X} g(x) P(X =x)
\end{equation}

Suppose $g(X) = X^2$, then

\begin{eqnarray*}
E(g(X)) &=&  \sum_{x \in X} g(x) P(X = x) \\
        &=&  \sum_{x \in X} x^2 P(X = x)
\end{eqnarray*}

***Example: Rolling One Die***  
Let a random variable $X$ be the resulting number. Compute $E(X^2)$.  
**Solution:**  
\begin{equation}

\end{equation}

## Scaling and Translation

Suppose $g(X) = a + bX$, then
\begin{eqnarray*}
E(g(X)) &=&  \sum_{x \in X} g(x) P(X = x) \\
        &=&  \sum_{x \in X} (a + bx) P(X = x) \\
        &=&  \sum_{x \in X} a P(X = x) + \sum_{x \in X} bx P(X = x) \\
        &=&  a \sum_{x \in X} P(X = x) + b \sum_{x \in X} x P(X = x) \\
        &=&  a  + b E(X)
\end{eqnarray*}

So, for example, if we add 4 to a random variable, then we add 4 to its expected value If we multiply that random variable by 3, then we multiply it's expected value by 3. 


## Linearity of Expectation

If $X_1, X_2, \ldots, X_n$ are random variables (independent or dependent), then $E(X_1 + X_2 + \ldots + X_n) = E(X_1) + E(X_2) + \ldots + E(X_n)$

This is a very important property and particularly strong since it is true irrespective of whether $X_i$'s are independent or dependent. This property along with Indicator Random variables is an important tool for sometimes solving difficult problems. We will illustrate this using two examples.


## Example: Aces in a Poker Hand
Given a 5 card hand, let $X$ denote the number of Aces. What is $E(X)$?

Let $X_j$ be an Indicator random variable that the $j$th card is an Ace. \begin{eqnarray*} E(X) &=& E(X_1 + X_2 + X_3 + X_4 + X_5) \\ &=& E(X_1) + E(X_2) + E(X_3) + E(X_4) + E(X_5) ~~~~~~~ \text{by Linearity of Expectation} \\ &=& 5 \times E(X_1) ~~~~~~~~~ \text{by Symmetry} \\ &=& \frac{5}{13} \end{eqnarray*}

Example: Expectation of an Indicator Variable in a Binomial Distribution
Let $X \sim Binom(n,p)$. We define $I_j$ to be the Indicator random variable denoting success in the $j$th trial. Thus $P(I_j = 1) = p$. We can now rewrite $X = I_1 + I_2 + \ldots + I_n$. Again using linearity of expectation and symmetry, we get $E(X) = E(I_1 + I_2 + \ldots + I_n) = n E(I_1) = np$.

## Example: Local Maxima

Consider a random permutation of numbers $1, 2, \ldots, n$ where $n \geq 2$. Find the expected number of local maxima.

Consider for example a sequence of 3, 2, 1, 4, 7, 5 , 6. For this 3 , 7 and and 6 are local maxima. Note that for the end points the comparison is made only with one adjacent number. For the other cases, comparisons is made with the two neighbors. Hence 7 is a local maxima.

Let $I_j$ be the Indicator random variable that denotes that the number in the $j$th position is a local maxima, $j \in 1,2, \ldots, n$. \begin{equation} I_j = \begin{cases} 0 & \text{if the $j$th position is not a maxima} \\ 1 & \text{if the $j$th position is a maxima} \end{cases} \end{equation}

\begin{eqnarray*} E(I_1 + I_2 + \ldots + I_n) &=& E(I_1) + E(I_2) + \ldots + E(I_n) \\ &=& \underbrace{2 \times \frac{1}{2}}_{\text{2 endpoints}} + \underbrace{(n-2) \times \frac{1}{3}}_{\text{(n-2) internal points}}\\ &=& \frac{n+1}{3} \end{eqnarray*}

## Moments and Variance

If  $X$ is the random variable, $E(X)$ is the expected value and  also called the first moment. We discussed above that  $E(X) = \sum_{x \in X} x P(X=x)$.

$E(X^2)$ is  called the second moment and is given by  $E(X^2) = \sum_{x \in X} x^2 P(X=x)$.  
The Variance of $X$ is given by:  
\begin{equation*}
  Var(X) = E((X - \mu)^2)
\end{equation*}
where $\mu = E(X)$. It can also be shown that
\begin{eqnarray*}
        Var(X)  &=&  E((X - \mu)^2)\\
                &=& E(X^2 - 2 \mu X + \mu^2) \\
                &=& E(X^2) - 2 \mu E(X) + \mu^2 \\
                &=& E(X^2) - (E(X))^2
\end{eqnarray*}

### Standard Deviation
The standard deviation of a random variable (denoted $\sigma$) is a measure of how "spread out" a function is, or the typical deviation from the mean. Mathematically, the standard deviation is simply:  
\begin{equation}
\sigma(X) = \sqrt{Var(X)}
\end{equation}  
It is measured in the same units as $X$ and has the forgivingly simple scaling property:  
\begin{equation}
\sigma(aX) = |a|\sigma(X)
\end{equation}

***Example: Rolling a single die***  
Let $X$ be the resulting number. Find the variance and standard deviation of $X$.  
**Answer:**  
We already know that $E(X) = 7/2$ and $E(X^2) = 91/6$, so  
\begin{equation}
Var(X) = E(X^2) - (E(X))^2 = \frac{91}{6} - \frac{49}{4} = \frac{105}{36} \approx 2.9166
\end{equation}

Taking the square root of the variance, we get $\sigma(X) = \sqrt{Var(x)} \approx 1.7078$.

We will consider more about standard deviations when we discuss the Standard Normal Distribution.

# Continous Random Variable 

The last discrete distribution we covered was the Poisson distribution. In that case, recall that we discussed the added complexity of dealing with intervals which could occur anywhere in the real values of time. We handled that case with the approximation of the Poisson pmf. Now, we enter fully into the set of real numbers, considering random variables which can take on any real number.

Like discrete random variables, continuous random variables are functions that map the outcomes of a sample space to the $\mathbb{R}$. This is shown in Figure \@ref(fig:continuous-random-variable). The difference is that the sample space now is uncountably infinite and they map to real values.


![continuous-random-variable](Figures/Continuous-random-variable.jpg)


Let $X$ denote the random variable and takes values $x \in \mathbb{R}$. Unlike the discrete case, we now have a problem in quantifying $P(X = x)$. In particular, $P(X = x)$ must be equal to  0. In fact, we can formally state:
\begin{equation}
P(X=a) = \int_a^a f(x)dx = 0
\end{equation}
This is because, even in an arbitrary small neighborhood of $x$ there are infinite number of values so the probability that $X$ $\text{exactly}$ takes the value $x$ will be 0.  Looking at it in another way if we assign a positive value  to $P(X = x)$, however small, since $x$ can take uncountably infinite number of values, then the sum of the probabilities will become infinite.

You might recall an earlier discussion of aleph null ($\aleph_0$), well that concept re-appears here, as with just about any discussion of the real numbers. Although the probability that the continuous random variable takes a specific value is 0 we can quite easily define the probability over a *range of values.* To make things concrete consider the temperature of Davis at 12 noon. Note that we are assuming that we have an measuring instrument that can measure the temperature with arbitrary precision.

![temperature-profile](Figures/Temperature-profile.jpg)

In order to work with continuous random variables, what will  make sense is to define probability for a range of values. In order to do that, we define a function $f_X (x)$ which we call the **probability density function (pdf)**. We first give  the following definition:  $f_X (x)$ is a function of $x$ that when multiplied by a small interval $\Delta x$ gives the probability that $X$ lies between $x$ and $x + \Delta x$.

\begin{equation}
	P(\text{Temperature lies in the range $x$ and $x + \Delta x$}) \approx f_X(x) \Delta x
\end{equation}

Note that the left hand side is a probability. $f_X (x)$ is not a probability; we get the probability when it is multiplied by a range $\Delta x$. Note that the right hand side gives an area.  Some observation:

1.  For a given value of temperature say $t = 95$ and $\Delta = 2$, the probability is  $0.005*2 = 0.01$. If we increase the $\Delta = 4$, then to probability is higher $0.005*4 = 0.02$. This makes sense since it is more likely for the temperature to fall in a larger range.

2. For the same range the probability is different for different values of the temperature. The probability that temperature is between 60 and 62 is $0.001 * 2 = 0.002$. Similarly, the probability that the temperature is between 82 and 84 is 0.014. The density is different at  different temperature values. The likelihood of the temperature being around 60 is low hence the probability density around 60 is low. The probability of the temperature being around 85 is high so the density around 85 is hig.  The probability over a fixed range is higher where the density is higher.   

The above definition is an approximation and is acceptable as long as the the change in the function over $\Delta x$ is small compared to the value of $f_X(x)$.

In order to extend the definition of probability to make it more rigorous, we consider the area under the curve instead of the rectangles. The more rigorous definition of pdf is as follows: $f_X (x)$ is a function of $x$ for which the area under the curve $f_X (x)$ between  $x$ and $x + \Delta x$ gives the probability that $X$ lies between $x$ and $x + \Delta x$.  In the above definition $\Delta x$ need not be small.


### Probability Density Function (PDF) 

If $X$ is a random variable, $f_X(x)$ will denote its density function, which has the following properties:

1. $f_X(x) \geq 0$ This clearly follows from the Axiom that probability is a positive quantity between 0 and 1.

2. $\int_{-\infty}^{\infty} f_X(x) dx  = 1$ This would be the probability of the entire sample space $S$.  

***Example:***

Suppose we have some Random Variable $X$, whose density function is given by:
\begin{equation}
f_X(x) = 
\begin{cases}
cx~~ \text{for} ~~0 \leq x \leq 2 \\
0~~otherwise
\end{cases}
\end{equation}
For what value of $c$ is $f_X(x)$ a valid density function?

**Solution:** 

We require that $f_X(x) \geq 0$ for $0 \leq x \leq 2 $. Hence, $c > 0$. Furthermore, from the properties of density functions, we know that:
\begin{equation}
\int_{-\infty}^{\infty}f_X(x)dx = 1
\end{equation}
So we will take the integral and set $c$ accordingly:
\begin{equation}
\int_{0}^2 cxdx = \frac{cx^2}{2} \bigg\rvert_0^2 = \frac{c4}{2} = 2c
\end{equation}
So it will be density function if $c=\frac{1}{2}$


### Cummulative Distribution Function (CDF) 
The definition of the CDF is the same as in the case  of the discrete random variable

\begin{eqnarray*}
F_X (y) &=& P(X \leq y) \\
        &=& \int_{-\infty}^y f_X(x) dx
\end{eqnarray*}

and similar to the case of the discrete random variable

\begin{eqnarray*}
P(a < X \leq b) &=& F_X (b) - F_X(a)  \\
                &=& \int_{a}^{b} f_X (x) dx
\end{eqnarray*}

In continuous random variables $<$ (or $>$) and $\leq$ ($\geq$) are the same since the probability  that the random variable takes a specific values is equal to 0.

### Mean and Variance

The expected value of a continuous random variable  $E(X)$ is given by

\begin{equation}
E(X) = \int_{-\infty}^{\infty} x f_X(x)dx
\end{equation}

and the Variance

\begin{equation}
\text{Var}(X) = \int_{-\infty}^{\infty} (x - E(X))^2 f_X(x) dx
\end{equation}

***Example:***  
$X$ is some continuous random variable with density function:
\begin{equation}
f_X(x) = \frac{1}{2}x~~0 \leq x \leq 2
\end{equation}
Find $E(X)$, $E(X^2)$, and $Var(X)$.

**Solution:**  
\begin{align}
E(X) &= \int_0^2 x\frac{1}{2}xdx = \frac{x^3}{6} \bigg\rvert_0^2 = \frac{8}{6} = \frac{4}{3} \\
E(X^2) &= \int_0^2 x^2\frac{1}{2}xdx = \frac{x^4}{8} \bigg\rvert_0^2 = 2 \\
Var(X) &= E(X^2) - E(X)^2 = 2 - \frac{16}{9} = \frac{2}{9}
\end{align}

## Exponential Distribution

A random variable $X$ is exponentially distributed $X \sim \text{Expo}(\lambda)$ where $\lambda > 0$ is called he rate parameter.  The pdf is given by

\begin{equation*}
      f_X (x) = \left \{ \begin{array} {ll}
      \lambda e^{- \lambda x} &  x \geq 0 \\
      0 & otherwise
      \end{array} \right.
\end{equation*}

***Example: Check that this is even a valid PDF***  
**Solution:**  
\begin{equation}
\int_0^\infty \lambda e ^{-\lambda x}dx = -e^{-\lambda x} \rvert_0^\infty = 0-(-1) = 1
\end{equation}

This happens to put us very close to finding the CDF of the exponential distribution, which results from taking the integral from 0 to $x$. In this case, $-e^{0}$ is still -1, and by the same double subtraction as above, the CDF becomes:

\begin{equation}
F_X(x) = 1-e^{-\lambda x}
\end{equation}

###  Properties

\begin{eqnarray*}
      E(X)  &=& \frac{1}{\lambda} \\
      Var(X) &=& \frac{1}{\lambda^2}
\end{eqnarray*}

### Memory-less property of Exponential distribution
Let $X \sim \text{Expo}(\lambda)$ and denote the time until the next event. Suppose, we have observed that time $s$ has passed and the next event has not yet occurred. We are interested in finding the remaining  time until the next event. Let $Y$ denote this random variable. We would like to find $F_Y (t)$ which by definition equal to $P(Y \leq  t)$.  We will find the complement $P(Y > t)$

\begin{eqnarray*}
    P(Y > t) &=& P(X > t+s | X > s)  \\
                  &=&  \frac{P(X > t+s  \cap  X > s)}{P(X > s) }~~~~\text{Definition of conditional probability}\\
                  &=&  \frac{P(X > t+s )}{P(X > s) }~~~~\text{Intersection is simply s} \\
                  &=& \frac{1 - (1 - e^{-\lambda (s+t)})}{1 - (1 - e^{- \lambda s})}~~~~\text{Still exponential} \\
                  &=& e^{-\lambda t}
\end{eqnarray*}

Which shows that distribution of $Y$ is the same as $X$.  The Exponential distribution is the only continous distribution that has the memory-less property.

***Example: Waiting for a Shared Resource***

Suppose duration of telephone calls $X$ is exponentially distributed with rate parameter $\lambda = \frac{1}{180}$ per second. Suppose you arrive at a telephone booth to find that  there is person already  using it. What is the probability that you have to wait more than 50~seconds before you get to use the booth?

**Solution:**  
First  even though the person was already using the phone booth when you arrived, the remaining time $Y$ which the person will use the booth is still $\textbf{Expo}(\lambda)$. This follows from the memory-less property.  Thus,

\begin{eqnarray*}
F_Y(t) &=& P(Y \leq t) \\
            &=& \int_{-\infty}^{t} f_X (x) dx \\
            &=& 1 - e^{- \lambda t}
\end{eqnarray*}

We need to find $P(Y > 50)$ which is given by

\begin{eqnarray*}
P(Y > 50) &=& 1 - P(Y \leq 50) \\
            &=& 1 - F_Y(50) \\
            &=& 1 - (1 - e^{- \lambda 50})  \\
            &=& e^{- \frac{50}{180} }
\end{eqnarray*}

**Solution:**
Let $X_1$ and $X_2$ be the random variable that denotes the execution time in core 1 and core 2, respectively. Since the new job will be started to be served as soon as any one of the jobs is finished, we can see that $W = \min(X_1 , X_2)$. We want to $P(W > 5)$


\begin{eqnarray*}

           &=& P(X_1 > 5 , X_2 > 5) \\
           &=& \underbrace{P(X_1 > 5) \times P(X_2 > 5)}_{\text{due to independence}} \\
           &=&  e^{-\frac{1}{5} \times 5} e^{-\frac{1}{4} \times 5} \\
           &=& e^{-\frac{9}{4}}
\end{eqnarray*}

## Normal Distribution 

This is an important distribution. By  the Central Limit Theorem, the sum of lots of independent and identically distributed (iid) random variables looks like a Normal distribution.  If $X$ has a Normal distribution, the we write $X ~\sim \text{Norm}(\mu, \sigma^2)$ where $\mu$ and $\sigma^2$ are the two parameters. The pdf of the distribution is given by

\begin{equation}
    	f_X (x) = \frac{1}{\sqrt{2 \pi} \sigma} e^{- \frac{(x - \mu)^2}{2 \sigma^2}} ~~~~~~~~ - \infty < x < \infty
\end{equation}


```r
x   <- seq(10, 130, 1) # Generate a sequence of integers from 0 to 120
y   <- dnorm(x,mean=70, sd=15)  # Calculate the density for the different value in x
plot(x,y, type="l", main = "Density of Normal Distribution", xlim=c(10,130), ylab = "PDF")
```

### Properties
The following are some of the properties of Normal random variable

1. The pdf is symmetrical around  $\mu$. If we consider two values $x_1 = \mu + t$ and $x_2 = \mu - t$ then you can show that $f_X(x_1) = f_X(x_2)$.

2. The maximum value of $f_X (x)$ occurs at $x = \mu$ and the maximum value is $\frac{1}{\sqrt{(2 \pi) \sigma}}$.

3. From the maximum value the function decays exponentially (very fast).



### Facts
The following are some facts about the Normal distribution. We will not prove them. Some of them are easy to prove while some are not.

1. $f_X (x)$ is a valid pdf. In order to prove this we need to show that a) $f_X (x) \geq 0$ and b) $\int_{-\infty}^{\infty} f_X (x) = 1$.

2. $E(X) = \mu$ and $Var(X) = \sigma^2$. Thus the parameters of the density function are the mean and the variance of the distribution.

3. $F_X (a) = P(X \leq x) = \int_{-\infty}^{a} f_X (x) dx$. This is by definition.

4. If $X \sim N(\mu, \sigma^2)$ and suppose $Y = b + aX$, where $a$ and $b$ are constants, then $Y \sim N(a \mu + b, a^2 \sigma^2)$.

5. There is a special Normal distribution $Z \sim N(0,1)$ which is called the Standard Normal. This is a Normal distribution with mean 0 and variance 1.

```r
x   <- seq(-3, 3, 0.01) # Generate a sequence of integers from -10 to 10 
y   <- pnorm(x,mean=0, sd=1)  # Calculate the CDF of x
plot(x,y, type="l", main = "CDF of Standard Normal Distribution", ylab = "CDF")
```

6. For $Z$ we will use $\Phi(x)$ to denote $P(Z \leq x)$. Notice that because of the symmetry property
\begin{equation}
\Phi(x) = 1 - \Phi(-x)
\end{equation}
This is useful ad hence Standard Normal table only need to be created for the positive values of $z$.

7. If $X ~\sim N(\mu, \sigma^2)$, then using Fact 3, $X = \mu + \sigma Z$. Using Fact 3  $X \sim N(\sigma * 0 + \mu, \sigma^2 * 1) = N(\mu, \sigma^2)$. Thus $\frac{X - \mu}{\sigma} = Z ~\sim N(0,1)$
             .
8. We can use the above fact to compute probabilities for any Normal distribution $X ~\sim N(\mu, \sigma^2)$.

\begin{eqnarray*}
	           F_X (a) &=& P(X \leq a) \\
	                   &=& P(\frac{X - \mu}{\sigma} \leq \frac{a - \mu}{\sigma}) \\
	                   &=& P(Z \leq \frac{a - \mu}{\sigma}) \\
	                   &=& \Phi(\frac{a - \mu}{\sigma})
\end{eqnarray*}





***Example: More Realistic Batch Job Run Times***  
Suppose the runtime of a job on the CORI supercomputer is a random variable $X$ which has a Normal distribution with mean $\mu = 3$ mins  and standard deviation $\sigma = 3$ mins, i.e., $\sigma^2 = 9$. Find the probability that the run time of the job will be between 2 and 5 mins.


**Solution:**  
\begin{eqnarray*}

	                   &=& \Phi(\frac{2}{3}) - (1 - \Phi(\frac{1}{3}) \\
	                   &=& \Phi(\frac{2}{3}) + \Phi(\frac{1}{3}) - 1 \\
	                   &=& 0.3779
\end{eqnarray*}

# Entropy

Claude E. Shannon published "A Mathematical Theory of Communication" in the Bell System Technical Journal in 1948. With this groundbreaking work came the dawn of the information age. This chapter focuses on just a small subset of this work, and shows a broad application of these concepts. We begin with the concept of Information Entropy. About a decade before Shannon's work, Morse Code had been widely proliferated. In Morse Code, the letter E is represented with just a single dot $\cdot$, while the letter Y is represented by -- $\cdot$ -- --. Why is this?

Indeed E is quite common, while Y is rare. Therefore, it makes practical sense to represent the E with a shorter *codeword*. A codeword is just an alternative method of representing information for transmission over some channel. Letters in the English alphabet are chosen with some probability, so they could be modeled with a stochastic process. Can we come up with some formal way to choose our codewords optimally? It is precisely this problem Claude Shannon was concerned with. The formalization of this problem is Shannon Entropy.

Suppose a random variable $X$ takes on values $x_1, x_2,  \ldots, x_k$ with probabilities $p_1, p_2, \ldots, p_k$. We define entropy $H(X)$ as follows
\begin{equation}
H(X) = - \sum_{i=1}^{n} p_i log_2(p_i)
\end{equation}
The log is taken with base 2, and outputs the amount of information contained in the stochastic process in bits. $H(X)$ is also called the Shannon entropy and is the foundation of information theory.

Suppose $X \sim \text{BINOM}(10, 0.5)$. We will write an R code to compute  the entropy.

```r
n = 50
p = 0.3
x = seq(0,n,1)
prob = dbinom(x, n, p)
entropy = -sum(prob * log2(prob))
print(entropy)
```

Now let's plot the entropy as we change $p$. 

```r
n = 10
p = c(seq(0.1, 0.9, 0.1))
entropy = numeric(length(p))
for (i in seq(1,length(p),1)){
  x = seq(0,n,1)
  prob = dbinom(x, n, p[i])
  entropy[i] = -sum(prob * log2(prob))
  print(entropy[i])
}
plot(p,entropy, type = "b", pch = 19, ylab="Entropy", xlab = "p")
```

Indeed, for a given value of $n$ the uncertainty is maximum wher $p=0.5$. You can easily show that that variance is maximized when $p = 0.5$. When $p$ is close to 0 or 1, the uncertainity is low (and indeed whn $p$ is 0 or 1, there is no uncertainty). Hence, at least in the binomial case, the maximum amount of information is conveyed when the symbol could be chosen any which way.

### Analysis

1. Tradeoff between sensitivity ($\eta$) and specificity ($\theta$) These are defined as

\begin{eqnarray*}
   \eta  &=& P(T=1|D=1) \\
    \theta &=& P(T=0|D=0)
\end{eqnarray*}

2. ROC (Receiver Operating Characteristic) is plot of **True Positive Rate** $\eta$  versus **False Positive Rate** $1 - \theta$. Note that:

\begin{eqnarray*}
   \eta  &=& P(T=1|D=1) \\
           &=& \frac{P(T=1 \cap D=1)}{P(D=1)} \\
   1-  \theta &=& 1 - P(T=0|D=0) \\
                    &=& \frac{P(T=1 \cap D=0)}{P(D=0)}
\end{eqnarray*}

  
3. Predictive value of a positive test $\gamma$. This is given by

\begin{eqnarray*}
    \gamma &=& P(D=1|T=1) \\
           &=& \frac{P(D=1 \cap  T=1)}{P(T=1)} \\
           &=& \frac{P(T=1|D=1) \times P(D=1)}{P(T=1)} \\
           &=& \frac{P(T=1|D=1) \times  P(D=1)}{P(T=1|D=1) \times P(D=1) + P(T=1|D=0) \times P(D=0) } \\
           &=& \frac{\pi \eta}{\pi \eta + (1 - \pi)(1 - \theta}
\end{eqnarray*}


4. Predictive value of a negative test $\delta$ can be similarly shown to be

\begin{eqnarray*}
    \delta &=& P(D=0|T=0) \\
           &=& \frac{P(D=0 \cap  T=0)}{P(T=0)} \\
           &=& \frac{P(T=0|D=0) \times P(D=0)}{P(T=0)} \\
           &=& \frac{P(T=0|D=0) \times P(D=0)}{P(T=0|D=0) \times P(D=0) + P(T=0|D=1) \times P(D=1) } \\
           &=& \frac{\theta \times (1 - \pi) }{\theta \times (1 - \pi) + \pi \times (1 - \eta)}
  \end{eqnarray*}


## Example 1

Suppose the prevalence ($\pi$) of a disease is 0.5\%. Suppose the sensitivity ($\eta$) of the test is 0.98 and the specificity ($\theta$) is 0.96. What is the predictive value of a positive test $\gamma$, i.e., if a random person tests +ve what is probability that the person has the disease. 

**Solution**

\begin{eqnarray*}
  \gamma &=& P(D=1|T=1) \\
           &=& \frac{\pi \eta}{\pi \eta + (1 - \pi)(1 - \theta}\\
           &=& \frac{0.98 \times 0.005}{0.98 \times 0.005 + 0.004 \times 0.995} \\
           &=& 0.1096
\end{eqnarray*}


## Example 2

Suppose that a detection test for a particular  malware  in code is to be applied to all the executable codes in a system. The goal is to detect the malware quickly so that it can be removed before it spread to other code blocks. We will adopt the following convention. The test is positive ($T=1$) if malware is detected. 3. The entire population of executable codes  can be viewed as having been split into four groups: true and false positive, true and false negative. What proportion of the entire population falls into each of these four categories?

**Solution**

\begin{eqnarray*}
    P(D=1|T=1) &=& \frac{P(D=1 \cap T=1)}{P(T=1)} \\
               &=& \frac{P(T=1|D=1)P(D=1)}{P(T=1)} \\
               &=& \frac{P(T=1|D=1)P(D=1)}{\underbrace{P(T=1|D=1)P(D=1) + P(T=1|D=0)P(D=0)}_{\text{Using the law of total probability}}} \\
               &=& \frac{0.98*0.005}{0.98*0.005 + (1-0.96)*0.995}\\
               &=& 0.1096197
  \end{eqnarray*}


2. For this part we need to find $P(T=1)$. This is the same as the denominator:

\begin{eqnarray*}
    P(T=1) &=& P(T=1|D=1)P(D=1) + P(T=1|D=0)P(D=0) \\
           &=& 0.98*0.005 + (1-0.96)*0.995 \\
           &=& 0.0447
\end{eqnarray*}
